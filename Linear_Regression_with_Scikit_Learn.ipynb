## Step 5. train a linear regression model

lr_model.fit(np.array(X_train).reshape(-1, 1),y_train)

1. We will convert the series X_train to numpy array.The reason for this reshaping is that scikit-learn expects the input features X to be a two-dimensional array (i.e., matrix-like shape) where each column is a feature and each row is an observation.

2. .reshape()
* .reshape() is a method used to change the shape of an array.
* The first argument, -1, is a placeholder that indicates the number of rows should be automatically determined based on the size of the array and the specified column.
* The second argument, 1, specifies that the resulting array should have a single column.

```python
X_train
```

```python
y_train
```

```python
# import linear regression value
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()

# train the lr model
lr_model.fit(np.array(X_train).reshape(-1, 1),y_train)


print("model is: {0} + {1}*x".format(lr_model.intercept_,lr_model.coef_[0]))


# evaluate on test set
# lr.score(X_test, y_test)
```

* After training the linear regression model, it will have learned the best-fit parameters (the intercept and coefficient) to model the relationship between X_train and y_train.

* lr_model.intercept_: This is the y-intercept of the linear regression equation. It represents the expected mean value of the target variable y when all input features X are zero.

* lr_model.coef_: This returns an array of coefficients, one for each feature. Since we're dealing with simple linear regression (with only one feature), we use lr_model.coef_[0] to get the coefficient of that single feature. This coefficient represents the change in the target variable for a one-unit change in the feature.

## Step 6. plot the model with the hypothesis

```python
# define x value (0,25)
x_plot = np.linspace(0,25 , 2)            # 2 data points between 0 ~ 25
x_plot = x_plot.reshape(-1, 1)             # preprocessing required by scikit-learn functions

# predict by the trained model
y_plot = lr_model.predict(x_plot)

# plot the results
plt.figure()
plt.plot(x_plot,y_plot,'r')
plt.plot(X_train,y_train,'o')
plt.title("{0} + {1}*x".format(lr_model.intercept_,lr_model.coef_[0]))
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

Based on this plot, we can observe that the model seems satisfactory. If the line were positioned below all the data points, it would suggest a potential issue.

## Step 7. Predict and Evaluate the Train Set Cost

```python
# predict the train set
y_train_pred = lr_model.predict(np.array(X_train).reshape(-1, 1))

# define a data frame with predictions and labels
df_train = pd.DataFrame({'X_train':X_train,'y_train':y_train,'y_train_pred':y_train_pred}).reset_index(drop=True)
df_train.head(10)
```

```python
# calculate cost
train_cost = sum(abs(df_train['y_train'] - df_train['y_train_pred'])**2)/(2*df_train.shape[0])
print('------------------------------')
print('train_cost = {0}'.format(train_cost))
print('------------------------------')
```

```python
df_train
```

## predict and evaluate the test set cost

```python
y_test_pred = lr_model.predict(np.array(X_test).reshape(-1, 1))

# define a data frame with predictions and labels
df_test = pd.DataFrame({'x_test':X_test,'y_test':y_test,'y_test_pred':y_test_pred}).reset_index(drop=True)
df_test.head(10)

# calculate cost
test_cost = sum(abs(df_test['y_test'] - df_test['y_test_pred'])**2)/(2*df_test.shape[0])
print('------------------------------')
print('test_cost = {0}'.format(test_cost))
print('------------------------------')


plt.figure()
plt.scatter(df_test['x_test'], df_test['y_test'])
plt.plot(x_plot,y_plot,'r')
plt.grid()
```

During the training phase, we obtained a train_cost value of 3.855985193076466. To determine whether a value of 3.8 is high or acceptable, it's essential to examine the specifics of the data. It's noteworthy that even though the test dataset and the train dataset appear similar, the test cost is approximately twice that of the train set. This suggests that our model might be underfitting or that there are inherent differences between the training and testing datasets that we need to consider.

# second order polynomial

create a new column 'x2' in the dataframe 'df'. This column contains the square of values from the 'x' column.

```python
df = pd.read_csv(path,names= ['x','y'])
df['x2'] = df['x'].pow(2)
df.head()
```

Rearrange the order of columns in a table

```python
df[['x','x2', 'y' ]]
```

```python
from sklearn.model_selection import train_test_split

X = df[['x','x2']]
y = df['y']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
# import linear regression value
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()

# train the lr model
lr_model.fit(np.array(X_train).reshape(-1, 2),y_train) # 2 becouse we want reshape for 2 variables x and  x2


print("model is: {0} + {1}*x + {2}*x^2".format(lr_model.intercept_,lr_model.coef_[0],lr_model.coef_[1]))


# evaluate on test set
# lr.score(X_test, y_test)
```

```python
# define x value (0, 25)
x_plot = np.linspace(0,25 , 100)            # 100 data points between 0 ~ 25
y_plot = lr_model.intercept_ + x_plot*lr_model.coef_[0] + lr_model.coef_[1]*np.power(x_plot,2)


# plot the results

plt.figure()
plt.subplot(1,2,1)
plt.plot(x_plot,y_plot,'r')
plt.plot(X_train['x'],y_train,'o')
plt.title('train')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.subplot(1,2,2)
plt.plot(x_plot,y_plot,'r')
plt.plot(X_test['x'],y_test,'o')
plt.title('test')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
y_train_pred = lr_model.predict(X_train)

# define a data frame with predictions and labels
df_test = pd.DataFrame({'x_test':X_train['x'].values,'y_train':y_train,'y_train_pred':y_train_pred}).reset_index(drop=True)
df_test.head(10)

# calculate cost
test_cost = sum(abs(df_test['y_train'] - df_test['y_train_pred'])**2)/(2*df_test.shape[0])
print('------------------------------')
print('train_cost = {0}'.format(test_cost))
print('------------------------------')
```

```python
y_test_pred = lr_model.predict(X_test)

# define a data frame with predictions and labels
df_test = pd.DataFrame({'x_test':X_test['x'].values,'y_test':y_test,'y_test_pred':y_test_pred}).reset_index(drop=True)
df_test.head(10)

# calculate cost
test_cost = sum(abs(df_test['y_test'] - df_test['y_test_pred'])**2)/(2*df_test.shape[0])
print('------------------------------')
print('test_cost = {0}'.format(test_cost))
print('------------------------------')


# plt.figure()
# plt.scatter(df_test['x_test'],df_test['y_test'])
# plt.plot(x_plot,y_plot,'r')
# plt.grid()
```
